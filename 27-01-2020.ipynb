## scatter plot

```python
plt.scatter([52,42,94,48,12,66],[14,28,95,55,58,95],color='g')
plt.title("scatter \nplot")
plt.show()
```

## area plot

```python
days=[1,2,3,4]
working=[8,8.5,9,10]
sleeping=[8,10,11,12]
plt.stackplot(days,working,sleeping,colors=['b','g'],labels=['working','sleeping'])
plt.legend()
plt.title("area \nplot")
plt.xlabel('days')
plt.ylabel('working,sleeping')
plt.show()
```

## pie chart

```python
days=[1,2,3,4]
working=[8,8.5,9,10]
sleeping=[8,10,11,12]
slices=[4,10,12]
variables=['days','working','sleeping']
plt.pie(slices,labels=variables,colors=['g','b','r'],startangle=180,autopct='%1.3f%%')
plt.show()
```
